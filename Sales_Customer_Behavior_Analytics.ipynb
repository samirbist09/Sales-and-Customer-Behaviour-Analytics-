```python
#PHASE 3: Data Cleaning
#Step 12: Remove Duplicates
customers.drop_duplicates(inplace=True)

orders.drop_duplicates(inplace=True)

order_items.drop_duplicates(inplace=True)

payments.drop_duplicates(inplace=True)

products.drop_duplicates(inplace=True)
```

```python
#Step 13: Convert Date Columns
orders['order_purchase_timestamp'] = pd.to_datetime(
    orders['order_purchase_timestamp']
)
```

```python
orders['order_delivered_customer_date'] = pd.to_datetime(
    orders['order_delivered_customer_date']
)
```

```python
#PHASE 4: Merge Datasets
#Step 14: Merge Customers + Orders
df = orders.merge(
    customers,
    on='customer_id',
    how='left'
)
```

```python
#Step 15: Merge Order Items
df = df.merge(
    order_items,
    on='order_id',
    how='left'
)
```

```python
#Step 16: Merge Payments
df = df.merge(
    payments,
    on='order_id',
    how='left'
)
```

```python
df = df.merge(
    products,
    on='product_id',
    how='left'
)
```

```python
#Step 1: Create Monthly Revenue Dataset
monthly_sales = (
    df.groupby(
        df['order_purchase_timestamp'].dt.month
    )['Revenue']
    .sum()
)

monthly_sales
```

```python
#Step 2: Prepare Data
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array(monthly_sales.index).reshape(-1,1)

y = monthly_sales.values
```

```python
#Step 3: Train Model
model = LinearRegression()

model.fit(X,y)
```

```python
#Step 4: Predict Next Month
future_month = np.array([[13]])

prediction = model.predict(future_month)

print(
    "Predicted Revenue for Next Month:",
    prediction[0]
)
```

```python
#Step 5: Plot Prediction
plt.figure(figsize=(8,5))

plt.scatter(X,y)

plt.plot(
    X,
    model.predict(X)
)

plt.title(
    'Sales Trend Prediction'
)

plt.xlabel('Month')
plt.ylabel('Revenue')

plt.show()
```
